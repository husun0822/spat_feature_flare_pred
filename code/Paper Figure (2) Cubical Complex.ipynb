```python
import h5py
import numpy as np
import pandas as pd
import gudhi
from datetime import datetime
import sys
import seaborn as sns
import copy
import matplotlib.pyplot as plt
import matplotlib
```

```python
f = h5py.File("../Mflare_data/M_flare_1h.hdf5","r")
v = f['HARP377']
frame = v['flare_2011.02.13_17:38:00']
br = np.array(frame['Br'])
PIL = np.array(frame['PIL_MASK'])
```

```python
cmap = sns.cubehelix_palette(hue=0.05, rot=0, light=0.95, dark=0, reverse=True, as_cmap=True) # configure a grayscale cbar

br_PIL = copy.deepcopy(br)
br_PIL[PIL==0] = 0
sns.heatmap(br_PIL, cmap = cmap, xticklabels=False, yticklabels=False)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (20,4))
matplotlib.rcParams.update({'font.size': 20})

br_1 = np.abs(copy.deepcopy(br_PIL))
br_1[np.abs(br_1) > 194.99] = 0.0
ax[0] = plt.subplot(1,3,1)
sns.heatmap(br_1, cmap = cmap, xticklabels=False, yticklabels=False)
ax[0].set_title("(a) Threshold = 194.99 G")

br_2 = np.abs(copy.deepcopy(br_PIL))
br_2[np.abs(br_2) > 583.35] = 0.0
ax[1] = plt.subplot(1,3,2)
sns.heatmap(br_2, cmap = cmap, xticklabels=False, yticklabels=False)
ax[1].set_title("(b) Threshold = 583.35 G")

br_3 = np.abs(copy.deepcopy(br_PIL))
br_3[np.abs(br_3) > 918.35] = 0.0
ax[2] = plt.subplot(1,3,3)
sns.heatmap(br_3, cmap = cmap, xticklabels=False, yticklabels=False)
ax[2].set_title("(c) Threshold = 918.35 G")
```

```python
fig.savefig("../paper_writing/threshold_example_br.png",dpi=300)
```

```python
br = np.array(frame['TOTUSJH'])
br_PIL = copy.deepcopy(br)
br_PIL[PIL==0] = 0
threshold = [3267.6,25500,62215]
unit = " G^2/m"

fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (20,4))

br_1 = np.abs(copy.deepcopy(br_PIL))
br_1[np.abs(br_1) > threshold[0]] = 0.0
ax[0] = plt.subplot(1,3,1)
sns.heatmap(br_1, cmap = cmap, xticklabels=False, yticklabels=False)
ax[0].set_title("Threshold = " + str(threshold[0]) + unit)

br_2 = np.abs(copy.deepcopy(br_PIL))
br_2[np.abs(br_2) > threshold[1]] = 0.0
ax[1] = plt.subplot(1,3,2)
sns.heatmap(br_2, cmap = cmap, xticklabels=False, yticklabels=False)
ax[1].set_title("Threshold = " + str(threshold[1]) + unit)

br_3 = np.abs(copy.deepcopy(br_PIL))
br_3[np.abs(br_3) > threshold[2]] = 0.0
ax[2] = plt.subplot(1,3,3)
sns.heatmap(br_3, cmap = cmap, xticklabels=False, yticklabels=False)
ax[2].set_title("Threshold = " + str(threshold[2]) + unit)
```
